```python
from google.colab import files
import os
import pandas as pd
import json
import fitz  # PyMuPDF for PDFs
from docx import Document

# Function to process TXT files
def process_text(file_name):
    with open(file_name, "r", encoding="utf-8") as f:
        return f.read()

# Function to process CSV files
def process_csv(file_name):
    df = pd.read_csv(file_name)
    return df.to_string()  # Convert entire CSV to a text representation

# Function to process JSON files
def process_json(file_name):
    with open(file_name, "r", encoding="utf-8") as f:
        data = json.load(f)
    return json.dumps(data, indent=2)  # Convert JSON to formatted text

# Function to process DOCX files
def process_docx(file_name):
    doc = Document(file_name)
    return "\n".join([para.text for para in doc.paragraphs])

# Function to process PDF files
def process_pdf(file_name):
    doc = fitz.open(file_name)
    return "\n".join([page.get_text() for page in doc])

# Upload files
uploaded = files.upload()

# Process each uploaded file and store its text content in the 'text' variable
text = ""

for file_name in uploaded.keys():
    file_extension = os.path.splitext(file_name)[1].lower()

    # Determine which function to run
    if file_extension == ".txt":
        text = process_text(file_name)
    elif file_extension == ".csv":
        text = process_csv(file_name)
    elif file_extension == ".json":
        text = process_json(file_name)
    elif file_extension == ".docx":
        text = process_docx(file_name)
    elif file_extension == ".pdf":
        text = process_pdf(file_name)
    else:
        text = f"⚠ Unsupported file type: {file_extension}. Please upload a valid file."

# Display extracted text (for debugging purposes)
print("\n📂 Extracted Content:\n")
print(text[:1000])  # Show only first 1000 characters
print("\n" + "="*80 + "\n")  # Separator for readability
```

```python
import re

def clean_text(text):
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces and line breaks
    text = re.sub(r'[\t\n\r\f\v]', ' ', text)  # Normalize whitespace
    return text.strip()

cleaned_text = clean_text(text)
```

```python
def remove_legal_citations(text):
    text = re.sub(r'\[\d+\]', '', text)  # Remove references like [1], [23]
    text = re.sub(r'\(\d{4}\)\s\w+', '', text)  # Remove case citations like (2020) XYZ v. ABC
    return text

cleaned_text = remove_legal_citations(cleaned_text)
```

```python

#using top_k
def split_into_chunks(text, max_words=400):
    words = text.split()
    return [" ".join(words[i:i+max_words]) for i in range(0, len(words), max_words)]

chunks = split_into_chunks(cleaned_text, max_words=400)

summaries = [
    pipe(chunk, max_length=100, min_length=20, do_sample=True, top_k=50, temperature=0.7)[0]['summary_text']
    for chunk in chunks
]

final_summary = " ".join(summaries)
print(final_summary)
```

```python
# Function to summarize text with top_k chunking
def summarize_text(text):
    cleaned_text = clean_text(text)
    cleaned_text = remove_legal_citations(cleaned_text)

    # Chunk the cleaned text into smaller parts
    chunks = split_into_chunks(cleaned_text, max_words=400)

    summaries = [
        summarizer(chunk, max_length=100, min_length=20, do_sample=True, top_k=50, temperature=0.7)[0]['summary_text']
        for chunk in chunks
    ]
    final_summary = " ".join(summaries)
    return final_summary

# Function to process raw text input (text-to-text)
def process_text_input(input_text)-> str:
    """Clean, remove citations, and summarize the raw text."""
    return summarize_text(input_text)

# Example of processing raw text and printing the summary
raw_text = input("Enter raw text: ")
summary = process_text_input(raw_text)
print("Summarized Text:")
print(summary)
```

```python
# Load Google Pegasus Summarization Model
summarizer = pipeline("summarization", model="google/pegasus-large")
```

```python
def summarize_text(text, chunk_size=512):
    words = text.split()
    max_len = min(len(words) // 2, 250)  # Half of original text, max 250
    min_len = min(len(words) // 3, 100)  # One-third of original text, min 100

    if len(words) < chunk_size:
        return summarizer(text, max_length=max_len, min_length=min_len, do_sample=False)[0]['summary_text']

    chunks = []
    current_chunk = []

    for word in words:
        current_chunk.append(word)
        if len(current_chunk) >= chunk_size:
            chunks.append(" ".join(current_chunk))
            current_chunk = []

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    summary = ""
    for chunk in chunks:
        summary_part = summarizer(chunk, max_length=max_len, min_length=min_len, do_sample=False)[0]['summary_text']
        summary += summary_part + " "

    return summary.strip()
```

```python
from transformers import pipeline

# Define Transport-related keywords
transport_keywords = [
    "shipment", "cargo", "freight", "transport", "logistics", "warehouse", "customs", "permit",
    "destination", "origin", "departure", "arrival", "carrier", "vessel", "bill of lading", "invoice",
    "hazardous", "flammable", "explosive", "toxic", "radioactive", "tax", "duty", "import", "export",
    "compliance", "safety", "documentation", "clearance", "insurance", "tracking", "consignment",
    "load", "unload", "vehicle", "fuel", "transit", "delay", "rail", "air", "marine", "route",
    "warehouse", "inventory", "distribution", "storage", "transportation", "movements", "fleet",
    "weigh station", "highway", "permit fees", "container", "supply chain", "overload", "fine",
    "border crossing", "inspection", "security", "seaport", "aviation", "regulation", "violation",
    "road tax", "custom duty", "tariff", "logistics hub"
]

# Load summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Extract relevant sentences
def extract_relevant_sentences(text, keywords):
    sentences = text.split(". ")
    relevant_sentences = [sent for sent in sentences if any(keyword in sent.lower() for keyword in keywords)]
    return " ".join(relevant_sentences)

# Chunk text for better accuracy
def chunk_text(text, max_words=500):
    words = text.split()
    return [" ".join(words[i:i+max_words]) for i in range(0, len(words), max_words)]

# Highlight keywords in text
def highlight_keywords(text, keywords):
    for keyword in keywords:
        text = re.sub(fr"\b({keyword})\b", r"\1**", text, flags=re.IGNORECASE)
    return text

# Summarization Function
def summarize_transport_text(text):
    relevant_text = extract_relevant_sentences(text, transport_keywords)
    if not relevant_text:
        return "No transport-related information found."

    chunks = chunk_text(relevant_text, max_words=500)
    summaries = [summarizer(chunk, max_length=200, min_length=100, do_sample=False)[0]['summary_text'] for chunk in chunks]

    summary = " ".join(summaries)
    return highlight_keywords(summary, transport_keywords)

final_summary = summarize_transport_text(cleaned_text)
print_summary = final_summary
print("\nGenerated Summary:")
print(final_summary)
```

```python
import gradio as gr
import pdfplumber
import pytesseract
import docx
from PIL import Image
import re
import nltk
from transformers import pipeline
from evaluate import load

# Download necessary resources
nltk.download("wordnet")

# Load AI Models
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
rouge = load("rouge")

# Define transport-related keywords
transport_keywords = [
    "shipment", "cargo", "freight", "transport", "logistics", "warehouse", "customs", "permit",
    "destination", "origin", "departure", "arrival", "carrier", "vessel", "bill of lading", "invoice",
    "hazardous", "flammable", "explosive", "toxic", "radioactive", "tax", "duty", "import", "export",
    "compliance", "safety", "documentation", "clearance", "insurance", "tracking", "consignment",
    "load", "unload", "vehicle", "fuel", "transit", "delay", "rail", "air", "marine", "route",
    "warehouse", "inventory", "distribution", "storage", "transportation", "movements", "fleet",
    "weigh station", "highway", "permit fees", "container", "supply chain", "overload", "fine",
    "border crossing", "inspection", "security", "seaport", "aviation", "regulation", "violation",
    "road tax", "custom duty", "tariff", "logistics hub", "shipment tracking", "driver compliance"
]

# Hazardous materials and their severity levels
hazardous_materials = {
    "flammable": 5,
    "explosive": 9,
    "radioactive": 10,
    "toxic": 8,
    "hazardous": 6,
    "chemical": 4,
    "corrosive": 7,
    "oxidizer": 6,
    "compressed gas": 5
}

# Extract text from different file types
def extract_text(file):
    if file.name.endswith(".pdf"):
        with pdfplumber.open(file.name) as pdf:
            return " ".join([page.extract_text() for page in pdf.pages if page.extract_text()])
    elif file.name.endswith(".docx"):
        doc = docx.Document(file.name)
        return " ".join([para.text for para in doc.paragraphs])
    elif file.name.endswith((".png", ".jpg", ".jpeg")):
        return pytesseract.image_to_string(Image.open(file.name))
    else:
        return "Unsupported file format."

# Clean extracted text
def clean_text(text):
    text = re.sub(r'\(cid:\d+\)', '', text)  # Remove encoding issues
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    text = text.replace("•", "-")  # Replace bullets
    return text.strip()

# Highlight keywords
def highlight_keywords(text, keywords):
    for keyword in keywords:
        text = re.sub(fr"\b({keyword})\b", r"\1**", text, flags=re.IGNORECASE)
    return text

# Calculate Disaster Severity Score
def calculate_disaster_severity(text):
    severity_score = 0
    found_hazards = []

    for material, score in hazardous_materials.items():
        if material in text.lower():
            found_hazards.append(material)
            severity_score += score

    if severity_score >= 15:
        severity_level = "*Extreme Danger: Immediate action required*"
    elif severity_score >= 10:
        severity_level = "*High Risk: Strict compliance needed*"
    elif severity_score >= 5:
        severity_level = "*Moderate Risk: Review and mitigate risks*"
    else:
        severity_level = "*Low Risk: No immediate danger detected*"

    return severity_level, found_hazards

# Generate Report on Transport & Risk Insights
def analyze_document(file):
    text = extract_text(file)
    if not text:
        return "Error: Could not extract text.", "Error: No text available to analyze.", "Error: No accuracy score."

    cleaned_text = clean_text(text)

    # Extract sentences related to logistics, transport, and compliance
    relevant_sentences = " ".join([sent for sent in cleaned_text.split(". ") if any(k in sent.lower() for k in transport_keywords)])

    if not relevant_sentences:
        return "No transport-related content found.", "No risks detected.", "No accuracy score."

    # Generate Risk Report
    hazard_keywords = ["hazardous", "flammable", "explosive", "radioactive", "toxic", "chemical"]
    expired_keywords = ["expired", "invalid", "not valid", "past due"]
    tax_keywords = ["tax", "duty", "gst", "tariff", "custom duty", "fine", "penalty"]
    compliance_keywords = ["violation", "non-compliant", "penalty", "inspection", "permit revoked", "license suspended"]
    security_keywords = ["smuggling", "unauthorized", "theft", "fraud", "breach", "tampering"]

    hazards = [word for word in hazard_keywords if word in cleaned_text.lower()]
    expired_docs = [word for word in expired_keywords if word in cleaned_text.lower()]
    tax_issues = [word for word in tax_keywords if word in cleaned_text.lower()]
    compliance_issues = [word for word in compliance_keywords if word in cleaned_text.lower()]
    security_issues = [word for word in security_keywords if word in cleaned_text.lower()]

    severity_level, found_hazards = calculate_disaster_severity(cleaned_text)

    risk_report = f"""
### *Key Transport & Compliance Insights*
- *Expired Documents:* {', '.join(expired_docs) if expired_docs else 'None'}
- *Tax Issues:* {', '.join(tax_issues) if tax_issues else 'None'}
- *Regulatory Compliance Issues:* {', '.join(compliance_issues) if compliance_issues else 'None'}
- *Security Concerns:* {', '.join(security_issues) if security_issues else 'None'}

### *Risk Report*
- *Hazardous Materials Identified:* {', '.join(hazards) if hazards else 'None'}
- *Disaster Severity Level:* {severity_level}
- *Potentially Dangerous Materials Present:* {', '.join(found_hazards) if found_hazards else 'None'}
"""

    return risk_report

# Q&A Function
def ask_question(file, question):
    text = extract_text(file)
    if not text:
        return "Error: Could not extract text."

    cleaned_text = clean_text(text)
    result = qa_pipeline(question=question, context=cleaned_text)
    return result["answer"]

# Create Gradio UI
with gr.Blocks() as demo:
    gr.Markdown("# 🚛 Transport Document Analyzer with Disaster Severity Report")

    file_input = gr.File(label="Upload a Transport Document")

    summarize_btn = gr.Button("Analyze Document")
    risk_output = gr.Markdown()

    question_input = gr.Textbox(label="Ask a Question About the Document")
    qa_btn = gr.Button("Get Answer")
    qa_output = gr.Markdown()

    summarize_btn.click(analyze_document, file_input, risk_output)
    qa_btn.click(ask_question, [file_input, question_input], qa_output)

demo.launch(share=True)
```
